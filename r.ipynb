#3. O que é uma Análise Exploratória de Dados (EDA)?

EDA é um processo antigo e vem sendo atualizado ao longo do tempo com uso de de diferentes técnicas e ferramentas. 

EDA é um processo de análise dados que em sua essência tem como objetivos:
 - descobrir padrões que estão escondidos nos dados; 
 - descobrir sobre a qualidade dos dados;
 - descobrir anomalias nos dados (dados faltantes, outliers e erros humanos)
 - entender melhor os dados (relacionamentos de variáveis);
 - entender melhor negócio que você está envolvido;
 - extrair variáveis importantes; 
 - excluir variáveis inutéis;
 - limpar o seu dataset. 

Em minha opinião, esse é o nucleo para a Ciência de Dados. Aqui é onde você pode se diferenciar dos demais. Por que? Porque se a EDA for bem feita o seu modelo irá ter um desempenho melhor, uma complexidade menor e, por consequência, mais barato no final. Como dizem na área "Se entrar lixo no seu modelo, ele vai te retornar lixo". 

## 3.1 Quais ferramentas matemáticas, estatíticas e de visualização utilizamos em uma EDA?


A ferramentas que utilizamos aqui para alcançar os objetivos citados acima são:
- **Técnicas de redução de dimensão dos dados**: alguns datasets possuem muitas variáveis e por isso é interessante reduzir a dimensão dele (diminuir as variáveis) para que você consiga ter uma análise qualitativa mais significativa; 
- **Técnicas de agrupamento:** mesmo intuito de cima, mas agrupando dados semelhantes por algum tipo de característica;
- **Níveis de Análise:**
  1.  **Análise e Visualização Univariada**: olhar para uma só variável e extrair o máximo de informações dela (distribuição, média, mediana etc.). É importante notar que aqui não vamos ter insights sobre a causa ou efeito de relacionamento com outras variáveis. 
    - Aqui você pode usar ferramentas de visualização para detecção de outliers (boxplots, violin plot);
    - Scatter plot de 1-D; 
    - Usar PDF e CDF para entender a distribuição e a frequência dos dados; 

  2.  **Análise e Visualização Bivariada**: olhar para duas variáveis (normalmente uma delas é a variável de interesse do seu projeto, mas você pode fazer entre cada variável também mas dois-a-dois) e extrair o máximo de informações sobre a relação delas.
    - Scatter plot de 2-D: ajuda a visualizar a correlação entre duas variáveis usando as coordenadas cartesianas; 
    - Pair plot: 

  3. **Análise e Visualização Multivariada**: mesmo conceito de cima, mas agora com pelo menos três ou mais variáveis ao mesmo tempo. O objetivo é entender a relação entre elas.  
    - Counter plot 

- **Estatística Descritiva:**
  - Estatísticas/Medidas de Resumo
    - Medidas de Localização ou Tendência Central
      - Média
      - Mediana
      - Moda
      - Interquartil
    - Medidas de Dispersão
      - Desvio Padrão
      - Amplitude (range)
      - Variância
      - Amplitude Interquatil

    - Medidas do Formato da Distribuição
      - Skewness
      - Kurtosis
    - Medidas de Dependência
      - Correlações

### 4.5.1 Estatística Descritiva

Vamos tentar tirar algumas informações de cada uma dessas variáveis. Primeiro, vamos dividir nossas variáveis em numéricas e categorias. 

```python
num_variables = df1.select_dtypes(include=['int64', 'float64'])
cat_variables = df1.select_dtypes(exclude=['int64', 'float64', 'datetime64[ns]'])
```

Vamos usar agora algumas operações para ter algumas medidas de tendencia e dispersão.

```python
def show_descriptive_statistical(df):
    # Central Tendency - mean, median
    ct1 = pd.DataFrame(df.apply(np.mean)).T
    ct2 = pd.DataFrame(df.apply(np.median)).T

    # Dispersion - std, min, max, range, skew, kurtosis
    d1 = pd.DataFrame(df.apply(np.std)).T
    d2 = pd.DataFrame(df.apply(min)).T
    d3 = pd.DataFrame(df.apply(max)).T
    d4 = pd.DataFrame(df.apply(lambda x: x.max() - x.min())).T
    d5 = pd.DataFrame(df.apply(lambda x: x.skew())).T
    d6 = pd.DataFrame(df.apply(lambda x: x.kurtosis())).T

    m = pd.concat([d2, d3, d4, ct1, ct2, d1, d5, d6]).T.reset_index()
    m.columns = ['attributes', 'min', 'max', 'range', 'mean', 'median', 'std', 'skew', 'kurtosis']
    print(m)
```

```python
show_descriptive_statistical(num_variables)
```

```python
## outra forma de fazer a mesma coisa, mas sem skew, kurtosis
num_variables.describe().apply(lambda y: y.apply(lambda x: format(x, 'f')))
```

### 4.5.2 Correlações (duas variáveis)

Tem pessoas que gostam de começar com a análise univariada aqui. Mas eu não sou ela. Eu gosto de começar a ter o entendimento maior do dataset e depois partir para os compomentos univariados. 

Por isso, vou começar a correlação. A correlação é uma medida que descreve o relacionamento entre duas variáveis. Ela pode ser negativa ou positiva.

```python
corr = df1.corr()

sns.heatmap(corr, 
            xticklabels=corr.columns, 
            yticklabels=corr.columns, 
            annot=True,
            cmap=sns.diverging_palette(220, 20, as_cmap=True))
```

Olhando para o odometro e o ano conseguimos ver uma correlação negativa. Ou seja, quanto maior o valor do odometro (km percorridos), menor é o ano do carro (mais velho). Esse é um simples exemplo. Podemos olhar pro preço e ano também. Quanto maior o ano, maior o preço do carro já que a sua correlação é positiva. 

### 4.5.3 Scatter Plot (duas variáveis)

Outra forma de visualizar o relacionamento entre duas variáveis são os gráficos de dispersão. É um gráfico que traça os valores de duas variáveis ao longo de dois eixos. É util para ver valores discrepantes.

Veja como os valores discrepantes (outliers) atrapalham a nossa visualização. Vamos fazer um filtro nesses valores e rodar a matrix e o scatter plot novamente.

### 4.5.4 Histogramas

Os histogramas são muitos bons para explorar uma única variável, ao contrário dos dois gráficos anteriores. É como se fosse um gráfico de barras, mas eles focam na distribuição de um CONJUNTO DE VALORES (bins) de uma variável.  

```python
## A maioria dos carros estem um hodômetro entre 0 e 200k. 
## Nomes técnicos aqui: a distribuição do hodõmetro é assimétrica positiva

df1['odometer'].plot(kind='hist', 
                     bins=50, 
                     figsize=(12,6), 
                     facecolor='grey',
                     edgecolor='black')
```

```python
# Os carros tem um ano em torno de 2000 a 2020
# Nomes técnicos aqui: o ano tem a distribuição assimétrica negativa
df1['year'].plot(kind='hist', 
                 bins=20, 
                 figsize=(12,6), 
                 facecolor='grey',
                 edgecolor='black')
```

É importante você entender essas distribuições de cabeças porque um numero grande de modelos de machine learning assumem que seus dados são normalmente distribuídos. Você acabou de ver ai em cima que não é verdade.
